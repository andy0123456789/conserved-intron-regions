```python
dtype=object

low_memory=False

dtype=object

df = pd.read_table('ccle/CCLE_RRBS_TSS_1kb_20180614.txt', sep='\t')


df = df.set_index('TSS_id')

df = df.drop( ['fpos' , 'chr', 'gene', 'tpos', 'strand', 'avg_coverage'], axis=1)

df = df.T

df = df.apply(pd.to_numeric, errors='coerce')
```

```python
keep_columns = df.columns[np.sum(np.isnan(df),axis=0) < 50]

df = df[keep_columns]

sns.distplot(np.sum(np.isnan(df),axis=0))
plt.show()
```

```python
df = df.fillna(df.mean(axis=0))

df.head(20)
```

```python
tss_stds = df.std(axis=0)

sns.distplot(tss_stds)
plt.show()

keep_tss = df.columns[tss_stds > 0.1]

variable_methyl = df[keep_tss]
```

```python
corrs = np.corrcoef(variable_methyl.T)
print(corrs.shape)
corrs
```

```python
sns.clustermap(corrs[:500,:500])
plt.show() 
```
